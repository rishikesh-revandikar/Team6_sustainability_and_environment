```python
# get the data from merged file of grid power
import pandas as pd

df= pd.read_csv('/content/drive/MyDrive/Hackwave/2019_to_2023_grid_all_columns.csv')
df
```

```python
# Convert the strings to datetime objects
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y %H:%M')

# Extract day, month, and time into separate columns
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['time'] = df['date'].dt.hour

df[]
```

```python
df.columns
```

```python
# select and make X df from Total Power (MW), Node 1 Power (MW), Node 2 Power (MW),Node 3 Power (MW), p1, p2, p3, c1, c2, c3

X_df = df[['month','day','time','Node 1 Power (MW)', 'Node 2 Power (MW)', 'Node 3 Power (MW)','p1','p2','p3' ,'c1', 'c2', 'c3']]
```

```python
X_df
```

```python
Y_df = df['stability']
Y_df
```

```python
# install sklearn of 1.3.0

!pip install scikit-learn==1.3.0
```

```python
# split X_df and Y_df in training and testing dataset

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
# make a model from random forest classifier with x_train and y_train

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=150,max_samples=20000,random_state=42)
model.fit(X_train, Y_train)
```

```python
#  find accuracy of model

from sklearn.metrics import accuracy_score
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python


# Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
gbc_model = GradientBoostingClassifier(random_state=42)
gbc_model.fit(X_train, Y_train)
gbc_pred = gbc_model.predict(X_test)
gbc_accuracy = accuracy_score(Y_test, gbc_pred)
print("Gradient Boosting Classifier Accuracy:", gbc_accuracy)

# Support Vector Machine Classifier
from sklearn.svm import SVC
svc_model = SVC(random_state=42)
svc_model.fit(X_train, Y_train)
svc_pred = svc_model.predict(X_test)
svc_accuracy = accuracy_score(Y_test, svc_pred)
print("Support Vector Machine Classifier Accuracy:", svc_accuracy)

# K-Nearest Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, Y_train)
knn_pred = knn_model.predict(X_test)
knn_accuracy = accuracy_score(Y_test, knn_pred)
print("K-Nearest Neighbors Classifier Accuracy:", knn_accuracy)

# AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier
abc_model = AdaBoostClassifier(random_state=42)
abc_model.fit(X_train, Y_train)
abc_pred = abc_model.predict(X_test)
abc_accuracy = accuracy_score(Y_test, abc_pred)
print("AdaBoost Classifier Accuracy:", abc_accuracy)
```
