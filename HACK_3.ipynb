```python
# get the data from merged file of wind power
import pandas as pd

df= pd.read_csv('/content/drive/MyDrive/Hackwave/merged_df_all.csv')
df
```

```python
#drop unrelavent columns

df = df.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
```

```python
df
```

```python
df.columns
```

```python
# distribute at each hour Power generated by system | (MW) with Node 1 will be taking 20% of total power, Node 2 will be
# taking 45% of total power and Node 3 will be taking 35% of total power generated
# respectively.

# Calculate the total power generated by the system for each hour
df['Total Power (MW)'] = df['Power generated by system | (MW)']

# Calculate the power taken by each node
df['Node 1 Power (MW)'] = df['Total Power (MW)'] * 0.20
df['Node 2 Power (MW)'] = df['Total Power (MW)'] * 0.45
df['Node 3 Power (MW)'] = df['Total Power (MW)'] * 0.35

# Print the updated DataFrame
df
```

```python
# rename datetime to date in df

df = df.rename(columns={"DateTime": "date"})
df
```

```python
#  make new index column

# Create a new index column
df = df.reset_index()
df = df.rename(columns={'index': 'new_index'})
df
```

```python
# Convert datetime format to the desired format
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%d-%m-%Y %H:%M')
df
```

```python
# Getting all price datasets and merging them
price_2019 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2019.csv')
price_2020 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2020.csv')
price_2021 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2021.csv')
price_2022 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2022.csv')
price_2023 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2023.csv')




```

```python
# concat above data frames of price

price_df = pd.concat([price_2019, price_2020, price_2021, price_2022, price_2023])
```

```python
# final price df
price_df
```

```python
#  make new index column

# Create a new index column
price_df = price_df.reset_index()
price_df = price_df.rename(columns={'index': 'new_index'})
price_df
```

```python
# Getting all consumtion datasets

consumtion_2019 = pd.read_csv('/content/drive/MyDrive/Hackwave/unit_consumption_2019.csv')
consumtion_2020 = pd.read_csv('/content/drive/MyDrive/Hackwave/unit_consumption_2020.csv')
consumtion_2021 = pd.read_csv('/content/drive/MyDrive/Hackwave/unit_consumption_2021.csv')
consumtion_2022 = pd.read_csv('/content/drive/MyDrive/Hackwave/unit_consumption_2022.csv')
consumtion_2023 = pd.read_csv('/content/drive/MyDrive/Hackwave/unit_consumption_2023.csv')

```

```python
#  concat above consumption df

consumtion_df = pd.concat([consumtion_2019, consumtion_2020, consumtion_2021, consumtion_2022, consumtion_2023])
```

```python
# final consumption df
consumtion_df
```

```python
#  make new index column

# Create a new index column
consumtion_df = consumtion_df.reset_index()
consumtion_df = consumtion_df.rename(columns={'index': 'new_index'})
consumtion_df
```

```python
# Getting stability datasets
stability_df = pd.read_csv('/content/drive/MyDrive/Hackwave/grid_stability_2019_2023.csv')

stability_df
```

```python
stability_df.isnull().sum()
```

```python
# remove 2184 rows from end in above stability_df to make proper dataset count for merging

stability_df = stability_df[:-2184]

stability_df
```

```python
#  make new index column

# Create a new index column
stability_df = stability_df.reset_index()
stability_df = stability_df.rename(columns={'index': 'new_index'})
stability_df
```

```python
# merge df , price_df , consumtion_dfand stability df on date

# Merge df, price_df, consumtion_df, and stability_df on 'date'
merged_df = df.merge(price_df, on='date')
merged_df = merged_df.merge(consumtion_df, on='date')
merged_df = merged_df.merge(stability_df, on='date')
```
