```python
# get the data from merged file of wind power
import pandas as pd

df= pd.read_csv('/content/drive/MyDrive/Hackwave/merged_df_all.csv')
df
```

```python
#drop unrelavent columns

df = df.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
```

```python
df
```

```python
df.columns
```

```python
# distribute at each hour Power generated by system | (MW) with Node 1 will be taking 20% of total power, Node 2 will be
# taking 45% of total power and Node 3 will be taking 35% of total power generated
# respectively.

# Calculate the total power generated by the system for each hour
df['Total Power (MW)'] = df['Power generated by system | (MW)']

# Calculate the power taken by each node
df['Node 1 Power (MW)'] = df['Total Power (MW)'] * 0.20
df['Node 2 Power (MW)'] = df['Total Power (MW)'] * 0.45
df['Node 3 Power (MW)'] = df['Total Power (MW)'] * 0.35

# Print the updated DataFrame
df
```

```python
# rename datetime to date in df

df = df.rename(columns={"DateTime": "date"})
df
```

```python
# prompt: make the date colum which is 2019-01-01 01:00:00.000	to 2019-01-01 01:00

df['date'] = df['date'].str.slice(0, 16)
df
```

```python
# Getting all price datasets and merging them
price_2019 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2019.csv')
price_2020 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2020.csv')
price_2021 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2021.csv')
price_2022 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2022.csv')
price_2023 = pd.read_csv('/content/drive/MyDrive/Hackwave/price_per_unit_2023.csv')




```

```python
# concat above data frames of price

price_df = pd.concat([price_2019, price_2020, price_2021, price_2022, price_2023])
```
