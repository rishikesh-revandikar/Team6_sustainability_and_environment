```python
# prompt: get the data from four xlxs file  and merge it acc to DateTime 2019
import pandas as pd

# Read the four Excel files
df1 = pd.read_excel('/content/drive/MyDrive/HACK/Hackathon_data/wind_power_data/air_temperature_2019.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/HACK/Hackathon_data/wind_power_data/power_gen_2019.xlsx')
df3 = pd.read_excel('/content/drive/MyDrive/HACK/Hackathon_data/wind_power_data/pressure_2019.xlsx')
df4 = pd.read_excel('/content/drive/MyDrive/HACK/Hackathon_data/wind_power_data/wind_speed_2019.xlsx')

# Merge the dataframes based on the DateTime column
merged_df = pd.merge(df1, df2, on='DateTime', how='outer')
merged_df = pd.merge(merged_df, df3, on='DateTime', how='outer')
merged_df = pd.merge(merged_df, df4, on='DateTime', how='outer')

# Print the merged dataframe
merged_df
```
